```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
plt.style.use('classic')
```

```python
a = 1
b = 2
bmax = 5
r = np.linspace(1e-10,bmax,1000) # arreglo no parte desde 0 para no calcular divisiones por 0 en el origen

def rhoi(a,b):
    return (3/(4*np.pi))/(b**3-a**3)

def rho(r,a,b):
    return 0*(r<a) + rhoi(a,b)*(r>=a)*(r<=b) + 0*(r>b)

def Emax(b):
    return 1/(b**2)

def E(r,a,b):
    return 0*(r<a) + ((r**3-a**3)/(r**2*(b**3-a**3)))*(r>=a)*(r<=b) + (1/(r**2))*(r>b)

def phi(r,a,b):
    phi1 = 3*(b**2-a**2)/(2*(b**3-a**3))
    phi2 = (-(2*a**3+r**3)/(2*r*(b**3-a**3)))+3*b**2/(2*(b**3-a**3))
    return phi1*(r<a) + phi2*(r>=a)*(r<=b) + (1/r)*(r>b)
```

# Densidad

```python
plt.figure()
plt.plot(r,rho(r,a,b), linewidth=2)
plt.xlabel('$r$', fontsize=15)
plt.title('Densidad $\\rho(r)$', fontsize=15)
plt.xticks([0,a,b],['$0$','$a$', '$b$'], fontsize=15)
plt.yticks([0,rhoi(a,b)],['$0$','$\\frac{Q}{4\\pi}\\frac{3}{(b^3-a^3)}$'], fontsize=15)
plt.ylim(-0.2*rhoi(a,b),1.2*rhoi(a,b))
plt.grid(True)
plt.tight_layout()
plt.savefig('fig-esfera-hueca-densidad.pdf')
```

```python
def grho(bma=1):
    plt.figure()
    plt.plot(r,rho(r,b-bma,b), linewidth=2)
    plt.xlabel('$r$', fontsize=15)
    plt.ylabel('$\\rho/Q$', fontsize=15)
    plt.title('Densidad $\\rho(r)$: a = '+str(np.round(b-bma,2))+', b = '+str(b), fontsize=15)
    plt.xlim(0,4)
    plt.ylim(0,0.5)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
interact(grho, bma=(0,b,0.01))
```

# Campo eléctrico

```python
plt.figure()
plt.plot(r,E(r,a,b), linewidth=2)
plt.xlabel('$r$', fontsize=15)
plt.title('Campo eléctrico $E(r)$', fontsize=15)
plt.xticks([0,a,b],['$0$','$a$', '$b$'], fontsize=15)
plt.yticks([0,Emax(b)],['$0$','$\\frac{Q}{4\\pi\\varepsilon_0 b^2}$'], fontsize=15)
plt.ylim(-0.2*Emax(b),1.2*Emax(b))
plt.grid(True)
plt.tight_layout()
plt.savefig('fig-esfera-hueca-campo.pdf')
```

```python
def gE(bma=1):
    plt.figure()
    plt.plot(r,E(r,b-bma,b), linewidth=2)
    plt.xlabel('$r$', fontsize=15)
    plt.ylabel('$(4\\pi\\varepsilon_0/Q)E(r)$', fontsize=15)
    plt.title('Campo eléctrico $E(r)$: a = '+str(np.round(b-bma,2))+', b = '+str(b), fontsize=15)
    plt.xlim(0,4)
    plt.ylim(-0.05,0.3)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
interact(gE, bma=(0.01,b,0.01))
```

# Potencial

```python
plt.figure()
plt.plot(r,phi(r,a,b), linewidth=2)
plt.xlabel('$r$', fontsize=15)
plt.title('Potencial $\\phi(r)$', fontsize=15)
plt.xticks([0,a,b],['$0$','$a$', '$b$'], fontsize=15)
plt.yticks([0,phi(a/2,a,b)],['$0$','$\\frac{Q}{4\\pi\\varepsilon_0}\\frac{3(b^2-a^2)}{2(b^3-a^3)}$'], fontsize=15)
plt.ylim(-0.2*phi(a/2,a,b),1.2*phi(a/2,a,b))
plt.grid(True)
plt.tight_layout()
plt.savefig('fig-esfera-hueca-potencial.pdf')
```

```python
def gphi(bma=1):    
    plt.figure()
    plt.plot(r,phi(r,b-bma,b), linewidth=2)
    plt.xlabel('$r$', fontsize=15)
    plt.ylabel('$(4\\pi\\varepsilon_0/Q)\\,\\phi(r)$', fontsize=15)
    plt.title('Potencial $\\phi(r)$: a = '+str(np.round(b-bma,2))+', b = '+str(b), fontsize=15)
    plt.xlim(0,4)
    plt.ylim(0,0.8)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
interact(gphi, bma=(0.01,b,0.01))
```
